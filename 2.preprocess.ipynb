```python
# not included: name, 
train_agg = train.groupby(['session_id', 'level_group']).agg({
    'index':'max',
    'elapsed_time':'max',
    'event_name' : 'count',
    'level':'mean',
    'room_coor_x':'nunique',
    'room_coor_y':'nunique',
    'screen_coor_x':'nunique',
    'screen_coor_y':'nunique',
    'hover_duration':'sum',
    'text':'nunique',
    'fqid':'nunique',
    'room_fqid':'nunique',
    'text_fqid':'nunique'
    }
)
train_agg
```

```python
# explode indexes
train_agg_idx = train_agg.reset_index()
train_agg_idx
```
